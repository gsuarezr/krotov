```python
import pruebas
import matplotlib.pyplot as plt
import numpy as np
```

```python
omega=1
ampl0=0.01
mu=0.1*omega
g=0.2*omega
T=np.pi/g
nt=100
tlist = np.linspace(0,T, nt)
gamma=0.05*omega
ratio,power,opt_result,energy,ergotropy=pruebas.get_result(omega,ampl0,mu,g,gamma,1,tlist,100,rate=1,energy=1+1.5j)
```

```python
t,energy2,ergotropy2=pruebas.giovanetti(omega,g,gamma,mu,1,T)
```

```python
ergotropy[-1]/energy[-1]
```

```python
ergotropy2[-1]/energy2[-1]
```

```python
pruebas.field_energy(opt_result,tlist,T,nt)
```

```python
2*np.pi/g
```

```python
t,energy2,ergotropy2=pruebas.giovanetti(omega,g,0.05,mu,1,T=500)
```

```python
pruebas.field_energy(opt_result,tlist,T,nt)
```

```python
plt.plot(tlist*g,np.real(energy),color='darkcyan')
plt.plot(tlist*g,np.real(ergotropy),ls='dashdot',color='darkcyan')
plt.plot(t*g,np.real(energy2),color='black')
plt.plot(t*g,np.real(ergotropy2),ls='dashdot',color='black')
plt.xlabel(r'g$\tau$',fontsize=14)
plt.ylabel(r'$\frac{E_{B}}{\omega}$,$\frac{\mathcal{E}_{B}}{\omega}$',fontsize=20)
plt.xlim(0,np.pi)
#plt.xticks(np.arange(0, 2*np.pi+0.001, np.pi/4),['0',r'$\frac{\pi}{4}$',r'$\frac{\pi}{2}$',r'$\frac{3 \pi}{4}$',r'$\pi$',r'$\frac{5\pi}{4}$',r'$\frac{3 \pi}{2}$',r'$\frac{7 \pi}{4}$',r'$2 \pi$'])
plt.show()
```
